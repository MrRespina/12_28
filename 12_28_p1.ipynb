```python
# merge 메소드를 이용해서 데이터 연결 > join
site = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/soldesk20231218/main/data/survey_site.csv')
person = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/soldesk20231218/main/data/survey_person.csv')
visited = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/soldesk20231218/main/data/survey_visited.csv')
survey = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/soldesk20231218/main/data/survey_survey.csv')

site,visited
df7 = site.merge(visited,left_on='name',right_on='site')
df7
```

```python
df8 = person.merge(survey,left_on='ident',right_on='person')
df8
```

```python
df8 = person.merge(survey,left_on='ident',right_on='person',how='left')
df8 # person의 join되지 않은 사람을 더 출력.
```

```python
df8 = person.merge(survey,left_on='ident',right_on='person',how='right')
df8 # person의 join되지 않은 사람을 더 출력.
```

```python
df8 = survey.merge(person,left_on='person',right_on='ident',how='right')
df8 # person의 join되지 않은 사람을 더 출력.
```

```python
ebola = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/soldesk20231218/main/data/country_timeseries.csv')
ebola

ebola.shape[0] - ebola.count() # 누락 값의 갯수.
print(np.count_nonzero(ebola.isnull())) # 누락 값의 총 갯수

# 총 데이터의 수
print(122*18 - np.count_nonzero(ebola.isnull()))

ebola['Cases_Liberia']
ebola['Cases_Liberia'].isnull().sum()
np.count_nonzero(ebola['Cases_Liberia'].isnull())
```

```python
# 누락값이 있는 경우 특정 값으로 채우기
### fillna() 라는 함수를 이용해서 누락값을 대체
ebola.iloc[0:10,0:5]
ebola.iloc[0:10,0:5].fillna(0)
ebola_fillna = ebola.fillna(0)
ebola_fillna

np.count_nonzero(ebola.isnull())
np.count_nonzero(ebola_fillna.isnull())

ebola_fillna_100 = ebola.iloc[0:10,0:5].fillna(100)
ebola_fillna_100
np.count_nonzero(ebola_fillna_100.isnull())
```

```python
# 누락 값을 평균 값으로 변경
ebolas = ebola.iloc[:,2:]
ebolas.mean()
ebolas.fillna(ebolas.mean())
ebolas.fillna(ebolas.mean()).iloc[:6,:6]
```

```python
# 빈 값이라면 바로 위 행에 있는 값으로 적용
ebola.fillna(method='ffill')

np.count_nonzero(ebola.fillna(method='ffill').isnull())
```

```python
# 아래 행에 있는 값으로 적용
ebola.fillna(method='bfill')

np.count_nonzero(ebola.fillna(method='bfill').isnull())
```

```python
## 위 아래의 중간 값으로 대체
ebola.interpolate()
np.count_nonzero(ebola.interpolate().isnull())
```

```python
# 누락값을 가진 행을 모두 삭제
ebola_drop = ebola.dropna()
ebola_drop

np.count_nonzero(ebola_drop.isnull())
```
